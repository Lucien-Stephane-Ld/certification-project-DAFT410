### Which are the healthiest cocktails (low sugar & low calories) ?

```python
top20_healthy = pd.read_csv('./CSVs/top20_healthiest.csv', sep=',')
```

```python
top20_healthy #only counting unique values for non-null rows & unique cocktails
```

```python
top20_healthy['sugar'] = top20_healthy['sugar'].str.replace('g', '').astype(float)
top20_healthy['calories'] = top20_healthy['calories'].str.replace('kcal', '').astype(float)
```

```python
top20_healthy['name'][2] = top20_healthy['name'][2].replace('Red Light, Green Light – Squid Game Shot', 'Red Light, Green Light')
```

```python
top20_healthy
```

```python
plt.style.use('fivethirtyeight')
sns.set(rc={"figure.figsize":(20, 7)})

chart_1 = sns.scatterplot(data=top20_healthy, x="name", y="calories", hue="sugar", size='sugar', sizes=(150,900))
chart_1.set_title('Healthiest Cocktails', fontdict={'size': 20, 'weight': 'bold'})

plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Cocktails', fontsize=18, fontdict={'weight':'bold'}, labelpad=30);
plt.ylabel('Calories count (kcal)', fontsize=18, fontdict={'weight':'bold'}, labelpad=20);

plt.legend(bbox_to_anchor=(1.02, 1), title='Sugar (g)', loc='upper left', fontsize=22)

plt.show()
```

### Which are the most popular ingredients ?

```python
top20_ingredients = pd.read_csv('./CSVs/top22_ingredients.csv', sep=',')
```

```python
top20_ingredients.head(20) #only counting unique cocktails & occurences for non-null rows
```

```python
top20_ingredients.dtypes
```

```python
top20_ingredients['occurrence'].astype(int)
```

```python
import squarify

values = top20_ingredients['occurrence'].astype(int)
ing_label = top20_ingredients['ingredient'].tolist()
occ_label = top20_ingredients['occurrence'].astype(int)

labels = [f"{ing_label}\n{occ_label}" for ing_label, occ_label in zip(ing_label, occ_label)]

squarify.plot(sizes = values,
              color = sns.color_palette("crest_r", len(values)),
              label = labels,
              text_kwargs = {'fontsize': 11, 'color': 'white', 'weight':'bold'},
              )
plt.title("Occurence of Ingredients",fontsize=20,fontweight="bold", pad=20)

plt.axis("off");
```

```python
sns.set(rc={"figure.figsize":(10, 7)});
# colors = sns.color_palette(palette='flare', as_cmap=True)

chart_3 = sns.scatterplot(data=top20_ingredients, x="occurrence", y="ingredient", hue="occurrence", size='occurrence', sizes=(150,900));
chart_3.set_title('Most Common Ingredients', fontdict={'size': 20, 'weight': 'bold'});

plt.xticks(fontsize=10);
plt.yticks(fontsize=10);

plt.xlabel('Cocktails', fontsize=18, fontdict={'weight':'bold'}, labelpad=30);
plt.ylabel('Calories count (kcal)', fontsize=18, fontdict={'weight':'bold'}, labelpad=20);

plt.legend([],[], frameon=False);
```

### Which cocktails have the most ingredients ?

```python
cocktail_density = pd.read_csv('./CSVs/cocktail_ingredient_density.csv', sep=',')
```

```python
cocktail_density_plot = cocktail_density
```

```python
cocktail_density_plot['ingredient_count'].value_counts()
```

```python
cocktail_density_plot
```

```python
cocktail_count = cocktail_density_plot.groupby('ingredient_count').size().reset_index(name='count')
cocktail_count
```

```python
plt.figure(figsize=(10, 7))

plt.bar(cocktail_count['ingredient_count'], cocktail_count['count'], color='#B572E6', alpha=1, width=1)

plt.title('Cocktail Count per Ingredient Count', fontsize=20, fontweight='bold', pad=20)

plt.xlabel('Ingredient Count', fontsize=18, fontweight='bold', labelpad=30)
plt.ylabel('Cocktail Count', fontsize=18, fontweight='bold', labelpad=20)

for i, row in cocktail_count.iterrows():
    count = row['count']
    plt.text(row['ingredient_count'], count, str(count), ha='center', va='bottom', fontsize=12)

plt.xticks( cocktail_count['ingredient_count'], cocktail_count['ingredient_count'], ha='center', fontsize=10)

plt.show() 
```

```python
plt.figure(figsize=(10, 7))

sns.histplot(data=cocktail_count, x='ingredient_count', weights='count', bins=len(cocktail_count), hue='count', alpha=1)

plt.title('Cocktail Count per Ingredient Count', fontsize=20, fontweight='bold')

plt.xlabel('Ingredient Count', fontsize=18, fontweight='bold', labelpad=30)
plt.ylabel('Count', fontsize=18, fontweight='bold', labelpad=20)

for i, row in cocktail_count.iterrows():
    count = row['count']
    plt.text(row['ingredient_count'], count, str(count), fontsize=12)

plt.xticks( cocktail_count['ingredient_count'], cocktail_count['ingredient_count'], ha='center', fontsize=10)

plt.legend().set_visible(False)

plt.show()
```

### Which letter is the most popular as first letter of a cocktail name?

```python
cocktail_first_letter = pd.read_csv('./CSVs/top_first_letter.csv', sep=',')
```

```python
cocktail_first_letter.head()
```

```python
sns.set(rc={"figure.figsize":(10, 7)});

chart_6 = sns.scatterplot(data=cocktail_first_letter, x="first_letter", y="count", hue="count", size='count', sizes=(150,900));
chart_6.set_title('Most Common First Letter', fontdict={'size': 20, 'weight': 'bold'});

plt.xticks(fontsize=10);
plt.yticks(fontsize=10);

plt.xlabel('Letter', fontsize=18, fontdict={'weight':'bold'}, labelpad=30);
plt.ylabel('Count', fontsize=18, fontdict={'weight':'bold'}, labelpad=20);

plt.legend([],[], frameon=False);
```

```python
df = cocktail_first_letter.sort_values('first_letter', ascending=False)
counts = cocktail_first_letter['count']

total_count = df['count'].sum()
df['percentage'] = (df['count'] / total_count) * 100

filtered_df = df[df['percentage'] > 3]
sorted_df = filtered_df.sort_values('first_letter')

colors = sns.cubehelix_palette( n_colors=len(sorted_df), start=0, rot=0, dark=.4, light=.75, reverse=True)

percentages = [(count / total_count) * 100 for count in counts]
filtered_labels = [f"{label} ({count})" for label, count in zip(sorted_df['first_letter'], sorted_df['count']) if (count / total_count) * 100 > 3]

plt.pie(sorted_df['count'], labels=filtered_labels, autopct='%1.1f%%', colors=colors, textprops={'color': 'black', 'weight': 'bold'}, startangle=90)

plt.title('First Letter Distribution', fontdict={'size': 16, 'weight': 'bold'}, pad=40)

plt.axis('equal') 
plt.tight_layout()
plt.show()
```

```python
df = cocktail_first_letter.sort_values('count', ascending=True)
counts = df['count']

total_count = df['count'].sum()
df['percentage'] = (df['count'] / total_count) * 100

filtered_df = df[df['percentage'] > 3]
sorted_df = filtered_df.sort_values('count')

colors = sns.cubehelix_palette(n_colors=len(sorted_df), start=0, rot=0, dark=.5, light=.95, reverse=False)

filtered_labels = [f"{label} ({count})" for label, count in zip(sorted_df['first_letter'], sorted_df['count']) if (count / total_count) * 100 > 3]

plt.pie(sorted_df['count'], labels=filtered_labels, autopct='%1.1f%%', colors=colors, textprops={'color': 'black', 'weight': 'bold'}, startangle=90)

plt.title('First Letter Distribution', fontdict={'size': 16, 'weight': 'bold'}, pad=40)

plt.axis('equal') 
plt.tight_layout()
plt.show()
```
